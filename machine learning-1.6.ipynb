```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('adult_income.csv', sep=',\s')
df.columns=['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race',
           'sex','capital_gain','capital_loss','hours_per_week','native_country','income']
df
```

```python
df.shape
```

```python
#cheaking the null value
df.isnull().values.any()
```

```python
#cheaking the information of dataset
df.info()
```

```python
#replacing the question mark with nan 
df=df.replace('?',np.nan)
df
```

```python
#cheaking the null value again.
df.isnull().values.any()
```

```python
#cheaking for null values
df.isnull().sum()
```

```python
#cheaking null value via visualisation
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

```python
#cheaking the information of the columns
df.info()
```

```python
#imputing the missing values with mode
for col in ['workclass','occupation','native_country']:
    df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
#now again cheaking for missing values.
df.isnull().values.any()
```

# NOW OUR DATA IS READY FOR VISUALIZATION WITHOUT ANY MISSING VALUES.

```python
#putting the data in pandas dataframe
df_income=pd.DataFrame(df)
df_income.head()
```

```python
#cheaking the various columns
df_income.columns
```

```python
#cheaking the data type of the columns
df_income.dtypes
```

FOR ANALYZING THE DATA WITH TARGET I.E INCOME WE HAVE TO CHANGE INCOME INTO NUMERIC TYPE.

```python
#changing the data type of target variable
df_income['income'] = df_income['income'].map({'<=50k':0, '>50k':1})
df_income['income'].dtypes
```
